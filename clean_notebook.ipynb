# üß† Text Summarization Using Pre-trained Models (Abstractive)

## üìò Description
This project demonstrates **Abstractive Text Summarization** using a **pre-trained Transformer model (BART)** from Hugging Face.  
It takes a **news article as input** and produces a **concise summary** automatically.

---

## üöÄ Features
- Uses **facebook/bart-large-cnn** pre-trained model  
- Generates human-like summaries  
- Supports **multi-line text input**  
- Saves each summary automatically in a `summaries/` folder  
- Lightweight and easy to run in VS Code  

---

## üß© Requirements
Install dependencies using:

```bash
pip install transformers torch


````python
# ---------------- IMPORT LIBRARIES ---------------------
from transformers import pipeline
from datasets import load_dataset
import evaluate
````

````python
# ---------------- LOAD DATASET ------------------------
print("üîπ Loading CNN/DailyMail dataset (small subset)...")
dataset = load_dataset("cnn_dailymail", "3.0.0", split="test[:1%]")  # 1% for testing
print(f"‚úÖ Dataset loaded successfully: {len(dataset)} samples.\n")
````

````python
# ---------------- LOAD PRE-TRAINED MODEL --------------
print("üîπ Loading BART summarization model...")
summarizer = pipeline("summarization", model="facebook/bart-large-cnn")
print("‚úÖ Model loaded successfully.\n")
````

````python
# ---------------- TEST ON SAMPLE ----------------------
sample_text = dataset[0]["article"]
reference_summary = dataset[0]["highlights"]

print("üì∞ Original Text (first 600 chars):\n")
print(sample_text[:600], "...\n")

print("üìò Reference Summary:\n")
print(reference_summary, "\n")

print("üîπ Generating model summary...")
summary = summarizer(
    sample_text,
    max_length=130,
    min_length=30,
    do_sample=False
)[0]["summary_text"]

print("üßæ Generated Summary:\n")
print(summary, "\n")
````

````python
# ---------------- EVALUATE USING ROUGE ----------------
print("üîπ Evaluating with ROUGE metrics...\n")
rouge = evaluate.load("rouge")

results = rouge.compute(
    predictions=[summary],
    references=[reference_summary]
)

print("‚úÖ ROUGE Scores:")
for k, v in results.items():
    print(f"  {k.upper()}: {v:.4f}")

print("\nüéØ Abstractive Text Summarization Completed Successfully!")
````

# üß† Interactive Text Summarizer using Pre-trained Model

## üìò Description
This project allows users to input any **news article or paragraph**, and it automatically generates a **concise summary** using a pre-trained **BART (facebook/bart-large-cnn)** model.

---

## ‚öôÔ∏è Features
- Accepts multi-line user input directly from the terminal  
- Uses **Abstractive Summarization** with the BART Transformer model  
- Automatically saves the generated summary to a text file for download  

---

## üß∞ Installation
Before running the program, install the required libraries in your **VS Code terminal**:

```bash
pip install transformers torch


````python


# ---------------- IMPORT LIBRARIES ---------------------
from transformers import pipeline
import datetime
import os

# ---------------- LOAD MODEL --------------------------
print("üîπ Loading BART model for summarization...")
summarizer = pipeline("summarization", model="facebook/bart-large-cnn")
print("‚úÖ Model loaded successfully!\n")

# ---------------- USER INPUT --------------------------
print("üì∞ Enter or paste your news/article text below.")
print("(Press ENTER twice when you are done):\n")

# Allow multiline input
lines = []
while True:
    line = input()
    if line == "":
        break
    lines.append(line)
text = " ".join(lines)

if not text.strip():
    print("‚ö†Ô∏è No text entered. Please try again.")
    exit()

print("\nüîπ Generating summary... Please wait...\n")

# ---------------- GENERATE SUMMARY --------------------
summary = summarizer(
    text,
    max_length=150,
    min_length=40,
    do_sample=False
)[0]["summary_text"]

# ---------------- DISPLAY SUMMARY ---------------------
print("üßæ Summary:\n")
print(summary)
print("\nüéØ Summarization completed successfully!\n")

# ---------------- SAVE SUMMARY TO FILE ----------------
timestamp = datetime.datetime.now().strftime("%Y-%m-%d_%H-%M-%S")
filename = f"summary_output_{timestamp}.txt"

# Create 'summaries' folder if it doesn't exist
os.makedirs("summaries", exist_ok=True)
filepath = os.path.join("summaries", filename)

with open(filepath, "w", encoding="utf-8") as f:
    f.write("==== ORIGINAL TEXT ====\n\n")
    f.write(text.strip() + "\n\n")
    f.write("==== GENERATED SUMMARY ====\n\n")
    f.write(summary.strip())

print(f"üíæ Summary saved successfully to: {filepath}")

````
